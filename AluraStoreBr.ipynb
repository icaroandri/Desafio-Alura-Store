### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python

faturamentos = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}


# Alterar o nome da variável para evitar sobrescrever o DataFrame 'loja'
for nome_loja, faturamento in faturamentos.items():
    print(f'O faturamento da {nome_loja} é de R${faturamento:,.2f}')

loja_maior_faturamento = max(faturamentos, key=faturamentos.get)

print(f'O maior faturamento foi da  {loja_maior_faturamento} em R$ {faturamentos[loja_maior_faturamento]:,.2f}')


import matplotlib.pyplot as plt

plt.clf()
plt.cla()
plt.close('all')

faturamentos = {
    'Loja 1': 1534509.12,
    'Loja 2': 1488459.06,
    'Loja 3': 1464025.03,
    'Loja 4': 1384497.58
}

chaves = list(faturamentos.keys())
valores = list(faturamentos.values())

plt.figure(figsize=(8,6))
plt.bar(chaves, valores)
plt.ylim(0, max(valores)*1.1)
plt.xlabel('Lojas')
plt.ylabel('Faturamento')
plt.title('Faturamento por Loja')

for i, v in enumerate(valores):
    plt.text(i, v, f'R$ {v:,.2f}', ha='center', va='bottom')

plt.show()

```

# 2. Vendas por Categoria


```python

print(type(loja))
print(type(loja2))
print(type(loja3))
print(type(loja4))


import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Agrupa por categoria e soma o preço
vendas_categoria = loja.groupby('Categoria do Produto')['Preço'].sum()

plt.figure(figsize=(12,6))
vendas_categoria.plot(kind='bar' , color = "yellow")

plt.xlabel('Categoria')
plt.ylabel('Faturamento')
plt.title('Faturamento por Categoria')

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Adiciona os valores em cima das barras
for i, v in enumerate(vendas_categoria):
    plt.text(i, v, f'R$ {v:,.2f}', ha='center', va='bottom')

plt.show()
print(vendas_categoria)
```

# 3. Média de Avaliação das Lojas

```python
medias_avaliacao = {
    'Loja 1': loja['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

for nome, media in medias_avaliacao.items():
    print(f'Média da avaliação da {nome}: {media:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos = loja['Produto'].value_counts()
produto_mais = produtos_mais_vendidos.idxmax()
produto_menos = produtos_mais_vendidos.idxmin()

print(f'Produto mais vendido: {produto_mais} ({produtos_mais_vendidos.max()} vendas)')
print(f'Produto menos vendido: {produto_menos} ({produtos_mais_vendidos.min()} vendas)')
```

```python
top_10 = produtos_mais_vendidos.head(10)

plt.figure(figsize=(12,6))
top_10.plot(kind='bar', color='orange')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xlabel('Produto')
plt.ylabel('Quantidade de Vendas')

for i, v in enumerate(top_10):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

# 5. Frete Médio por Loja


```python
fretes_medios = {
    'Loja 1': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

for nome, frete in fretes_medios.items():
    print(f'Frete médio da {nome}: R$ {frete:.2f}')
```

```python
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import zipfile
import urllib.request
import os

# Novo link do GitHub com shapefile do mundo
url = "https://github.com/nvkelso/natural-earth-vector/archive/refs/heads/master.zip"
zip_path = "mapa.zip"

urllib.request.urlretrieve(url, zip_path)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("mapa_mundo")

shapefile_path = "mapa_mundo/natural-earth-vector-master/110m_cultural/ne_110m_admin_0_countries.shp"

world = gpd.read_file(shapefile_path)

brasil = world[world['NAME'] == 'Brazil']

todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

plt.figure(figsize=(12,10))
brasil.plot(ax=plt.gca(), color='white', edgecolor='black')

sns.kdeplot(
    x=todas_lojas['lon'],
    y=todas_lojas['lat'],
    cmap="Reds",
    shade=True,
    bw_adjust=0.5,
    thresh=0.05,
    alpha=0.6
)

plt.title('Densidade de Vendas no Brasil')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
